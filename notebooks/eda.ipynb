```python
import pandas as pd
from pathlib import Path
```

```python
# chargement des fichiers CSV

# répertoire des données brutes :
data_dir = Path("..") / "data" / "raw"
csv_files = sorted([p for p in data_dir.iterdir() if p.suffix.lower() in [".csv", ".xlsx"]])

# liste de DataFrame (même ordre que csv_files) et dict {stem: df} :
dfs = [pd.read_csv(p) if p.suffix.lower() == ".csv" else pd.read_excel(p) for p in csv_files]
dfs_names = [p.stem for p in csv_files]
dfs_dict = dict(zip(dfs_names, dfs))
```

```python
# Afficher un aperçu de chaque DataFrame
for name, df in dfs_dict.items():
    print(f"--- {name} ---")
    display(df.head())
```

```python
# Taille des DataFrames
for name, df in dfs_dict.items():
    print(f" - {name}: {df.shape}")
```

```python
#info sur les DataFrames
for name, df in dfs_dict.items():
    print(f"--- Info for {name} ---")
    df.info()
```

```python
#aficher les valeurs manquantes
for name, df in dfs_dict.items():
    print(f"--- Missing Values for {name} ---")
    print(df.isnull().sum())
```

```python
# Afficher les types de données 
for name, df in dfs_dict.items():
    print(f"--- Data Types for {name} ---")
    print(df.dtypes)
```

```python
#aficher les douplicatas
for name, df in dfs_dict.items():
    print(f"--- Duplicates for {name} ---")
    print(df.duplicated().sum())
```

```python
#desrcibe des DataFrames
for name, df in dfs_dict.items():
    print(f"--- Description for {name} ---")
    display(df.describe(include='all'))
```

```python
#afficher les noms de colonnes
for name, df in dfs_dict.items():
    print(f"--- Columns for {name} ---")
    print(df.columns.tolist()) 
```

# Nettoyage

```python
#filtrer les données pour ne garder que les données à partir de septembre 2019 et sauvegarder en CSV
cta_df = dfs_dict["cta-ridership-daily-boarding-totals-20260203-69820a3f9df63091665572 (1)"]
cta_df = cta_df[cta_df["service_date"] >= "2019-09-01"]  
cta_df.to_csv("../data/raw/cta_ridership_daily_boarding_totals_2019_2025.csv", index=False)
```
